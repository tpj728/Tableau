# CitiBikes Data File - Combining, Cleaning, and Output for Tableau

#### Import Dependencies and create Trip and File Lists to Loop Through

```python
import pandas as pd
import numpy as np

#combine csv files
trip_csv = ['201712',
            '201801',
            '201802',
            '201803',
            '201804',
            '201805',
            '201806',
            '201807',
            '201808',
            '201809',
            '201810',
            '201811',
            '201812',
            '201901',
            '201902',
            '201903',
            '201904',
            '201905']
files = []

for month in trip_csv:
    files.append('JC-'+month+'-citibike-tripdata.csv')

print(files)

membership_csv = ['cZMp8',  #1/1/19
                 'FKjZg', #7/1/18
                 'MM5kM', #4/1/19
                 'pRAtm', #10/1/18
                 'tqNn6', #4/1/18
                 '7OLX8', #1/1/18
                 'K8j0z']  #10/1/17

files_m =[]
for file in membership_csv:
    files_m.append('data-'+file+'.csv')

print(files_m)
```

```python
len(files)
```

```python
data_files = []
for item in range(len(files)):
    data = 'data'+str(item+1) 
    data = pd.read_csv(files[item])
    data_files.append(data)
    print(data)
```

#### Append Tables together

```python
df = data_files[0]
for item in data_files[1:]:
    df = df.append(item)
```

#### Map Male, Female and Unknown Genders

```python
df['gender'] = df['gender'].map({0:'Unknown',1:'Male',2:'Female'})
```

#### Combine Files for Membership into a List

```python
membership_files = []
for x in range(len(files_m)):
    member = 'data'+str(x+1)
    m_file = files_m[x]
    member = pd.read_csv('../Membership Data/' + m_file)
    membership_files.append(member)
    print(membership_files)
```

#### Append items in Membership to a Consolidated DF

```python
membership_file = []
for y in range(len(membership_files)):
    mdata = membership_files[y]
    membership_file.append(mdata)
```

```python
dfm = membership_file[0]
for members in membership_file[1:7]:
    dfm = dfm.append(members)
```

#### Drop Arbitrary Column
